```python
import glob
from rdkit import Chem
from rdkit.Chem import Descriptors
import pandas as pd
```

```python
SMILES_molecules = pd.read_csv('OPEP2_dataset_SMILES.csv')
SMILES_molecules
```

```python
for x in range(len(SMILES_molecules)):
    mol = Chem.MolFromSmiles(SMILES_molecules.iloc[x][4])
    rad_electrons = Descriptors.NumRadicalElectrons(mol)
    
    if rad_electrons > 0:
        print(x)        
```
