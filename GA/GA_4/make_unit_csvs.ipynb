```python
import csv
import pandas as pd
from rdkit import Chem
```

```python
cores = pd.read_csv('../../building_blocks/cores_with_ringcount.csv', index_col=0)
unfused_cores = cores[cores['fused_ring_count'] < 4]

unfused_cores
```

```python
with open('GA4_donor_units.csv', mode='w+') as quick:
    quick_writer = csv.writer(quick)
    quick_writer.writerow(['index', 'smiles', 'filename_or_index'])

canonical_smiles_list = []
count = 0

for x in range(len(unfused_cores)):
    smiles = unfused_cores['smiles'].iloc[x]
    canonical_smiles = Chem.MolToSmiles(Chem.MolFromSmiles(smiles))
    
    if canonical_smiles not in canonical_smiles_list:
        with open('GA4_donor_units.csv', 'a') as quick_file:
            # write to quick analysis file
            quick_writer = csv.writer(quick_file)
            smi_filename = unfused_cores['smi_filename'].iloc[x]
            quick_writer.writerow([count, smiles, smi_filename])
            
        count +=1
        canonical_smiles_list.append(canonical_smiles)
        
with open('1235MonomerList.txt', "r") as file:
    Lines = file.readlines()
    line_index = 0
    for line in Lines:
        smiles = line.strip()
        canonical_smiles = Chem.MolToSmiles(Chem.MolFromSmiles(smiles))
        
        if canonical_smiles not in canonical_smiles_list:
            with open('GA4_donor_units.csv', 'a') as quick_file:
                # write to quick analysis file
                quick_writer = csv.writer(quick_file)
                quick_writer.writerow([count, smiles, line_index])

            count +=1
            canonical_smiles_list.append(canonical_smiles)
            
        line_index +=1
```
