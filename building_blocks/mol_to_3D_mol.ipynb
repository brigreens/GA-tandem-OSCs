# Convert ChemDraw SDF to 3D SDF file using RDKit

```python
from rdkit.Chem import AllChem as Chem
import glob
from rdkit.Chem import PeriodicTable
from rdkit.Chem import Descriptors
```

```python
def gen3d_rdkit(filename, directory):

    molecule = filename.split('\\')[-1].split('.')[0]
    print(molecule)

    m = Chem.MolFromMolFile(filename)
    
    smiles = Chem.MolToSmiles(m)
    
    m2 = Chem.AddHs(m)
    
    Chem.EmbedMolecule(m2,randomSeed=0xf00d, useRandomCoords=True)

    if '[se]' in smiles:
        Chem.MMFFOptimizeMolecule(m2)
    else:
        Chem.UFFOptimizeMolecule(m2)
    
    #PeriodicTable.GetAtomicSymbol
    
    # use MMFF for Se containing molecules
    #Chem.MMFFOptimizeMolecule(m2)

    print(Descriptors.NumRadicalElectrons(m2))

    new_filename = directory + molecule + '.mol'

    print(Chem.MolToMolBlock(m2),file=open(new_filename,'w+'))
```

```python
for file in glob.iglob('fragment_mols/*.mol'):

    gen3d_rdkit(file, 'fragment_mols/ff_optimized/')
```
