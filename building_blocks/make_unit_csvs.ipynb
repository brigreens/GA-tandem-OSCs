```python
import glob
import pandas as pd
```

```python
for smi in glob.iglob('fragment_mols/SMILES/*.smi'):
    with open(smi, 'r', encoding = 'utf-8') as file:
            smiles = file.readline()
            smiles = smiles.split('\t')[0]
            
    if '_TL' in smi:
        print('term_left')
    elif '_TR' in smi:
        print('term_right')
    else:
        print(smiles)


```
